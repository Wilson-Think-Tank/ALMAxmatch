### Now we are going to query by region (rather than target lists)

```python
from astropy import coordinates
from astropy import units as u

# sky coordinated of the vigo cluster
virgo_cluster = coordinates.SkyCoord('12h27m0s','12d43m0s',frame='icrs')
```

```python
# set the search region
myRegionSearch = archiveSearch(virgo_cluster)
myRegionSearch.targets
```

```python
# Run queries on target region with radius 6 deg (~ 1 virial rad, Hoffmann+1980, Kim+2014, https://arxiv.org/pdf/1409.3283.pdf)
myRegionSearch.runQueries(virgo_cluster, radius = 6*u.deg, region=True)

 # remove column that is long and not needed for now
myRegionSearch.queryResults[virgo_cluster].remove_column('Proposal authors')
```

```python
# make a human readable frequency column
myRegionSearch.parseFrequencyRanges()
```

```python
myRegionSearch.queryResults[virgo_cluster]
```

```python
from astropy.table import unique

sources = unique(myRegionSearch.queryResults[virgo_cluster], keys=['Source name'])[['Source name']]
print("number of unique sources:", len(sources))
```

## Now we match with SIMBAD to find those galaxies that are in the archive

```python
from astroquery.simbad import Simbad
result_table = Simbad.query_object(sources)
print(result_table)
```
